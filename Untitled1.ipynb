```python
import pandas as pd
```

```python
df = pd.read_csv('City_of_Burlington_Property_Details.csv',usecols=['StreetNumber','StreetName','Unit','LandUse','CurrentAcres',
                                                               'TotalGrossArea','FinishedArea','CurrentValue','CurrentLandValue',
                                                               'CurrentYardItemsValue','CurrentBuildingValue','BuildingType',
                                                               'HeatFuel','HeatType','Grade','YearBlt','SalePrice','NumofRooms',
                                                               'NumofBedrooms','NumofUnits','ZoningCode','Foundation','Depreciation',
                                                               'PropertyCenterPoint','FID'])
```

```python
df = df.set_index('FID')
```

```python
df.isnull().sum()/df.shape[0]
```

```python
# 'Unit' misses over 80% of datas, the 'Unit' column should be dropped
df = df.drop(columns=['Unit'])
```

```python
df.isnull().sum()/df.shape[0]
```

```python
#It is ambiguous whether a StreetNumber serves as a tag for a Street, or indicates a certain location on the street.
#The following shows that StreetNumber is Not a tag for a Street.
df_street=df[['StreetNumber','StreetName']].set_index('StreetNumber').sort_index()
print(df_street)
```

```python
df=df[df.StreetNumber.notnull()] # Drop the datas where the type of a building cannot be identified, losing 0.046% of datas
```

```python
df.isnull().sum()/df.shape[0]
```

```python
StreetNameCode = df['StreetName'].value_counts().to_dict()
df['StreetName']=df['StreetName'].map(StreetNameCode)
df
```

```python
LandUseCode = df['LandUse'].value_counts().to_dict()
df['LandUse']=df['LandUse'].map(LandUseCode)
df
```

```python
df=df[df.BuildingType.notnull()] # Drop the datas where the type of a building cannot be identified, losing 3.7% of datas
```

```python
df.isnull().sum()/df.shape[0]
```

```python
BuildingTypeCode = df['BuildingType'].value_counts().to_dict()
df['BuildingType']=df['BuildingType'].map(BuildingTypeCode)
df
```

```python
df=df[df.HeatFuel.notnull()] # Drop the datas where HeatFuel cannot be identified, losing 0.88% of datas
df.isnull().sum()/df.shape[0]
```

```python
HeatFuelCode = df['HeatFuel'].value_counts().to_dict()
df['HeatFuel']=df['HeatFuel'].map(HeatFuelCode)
df.HeatFuel
```

```python
df=df[df.HeatType.notnull()] # Drop the datas where HeatType cannot be identified, losing 0.88% of datas
df.isnull().sum()/df.shape[0]
```

```python
HeatTypeCode = df['HeatType'].value_counts().to_dict()
df['HeatType']=df['HeatType'].map(HeatTypeCode)
df.HeatType
```

```python
df=df[df.ZoningCode.notnull()] # Drop the datas where ZoningCode cannot be identified, losing 0.42% of datas
df.isnull().sum()/df.shape[0]
```

```python
ZoningCodeDict = df['ZoningCode'].value_counts().to_dict()
df['ZoningCode']=df['ZoningCode'].map(ZoningCodeDict)
df.ZoningCode
```

```python
df=df[df.Foundation.notnull()] # Drop the datas where the type of a building cannot be identified, losing 0.44% of datas
df.isnull().sum()/df.shape[0]
```

```python
FoundationCode = df['Foundation'].value_counts().to_dict()
df['Foundation']=df['Foundation'].map(FoundationCode)
df.Foundation
```

```python
df=df[df.PropertyCenterPoint.notnull()] # Drop the datas where PropertyCenterPoint cannot be identified, losing 1.3% of datas
df.isnull().sum()/df.shape[0]
```

```python
coordinates = [eval(item) for item in df['PropertyCenterPoint']]
x,y = zip(*coordinates)
x_coordinate = pd.DataFrame(list(x))
y_coordinate = pd.DataFrame(list(y))
```

```python
df
```

```python
df=df.reset_index()
```

```python
print(df.head())
```

```python
df = pd.concat([df,x_coordinate], axis=1)
df.columns.values[-1] = "PropertyCenterPoint_x"
df
```

```python
df = pd.concat([df,y_coordinate], axis=1)
df.columns.values[-1] = "PropertyCenterPoint_y"
df
```

```python
#df = df.drop(columns=['PropertyCenterPoint'])
df.to_csv('Property_2018.11.17.csv',index=False)
```
